# Business Understanding

The Environmental Protection Agency (EPA), National Aeronautics and Space Administration (NASA), and National Oceanic and Atmospheric Administration (NOAA) have come together in a collaborative effort to predict the presence of Cyanobacteria in freshwater sources.

Cyanobacteria, a harmful algal bloom (HAB), can have a severe impact on water quality and cause significant harm to the ecosystem, humans, pets, and marine life. HABs have the potential to:

- Disrupt ecosystems by hindering sunlight and oxygen
- Produce toxins that are hazardous to human and animal health
- Transfer these toxins to other aquatic creatures, posing a threat to aquaculture and agriculture

This analysis aims to demonstrate that Cyanobacteria severity levels can be predicted with minimal error by utilizing data already collected by government agencies. If successful, this proof of concept will secure further funding to make the predictive model accessible to local governments, helping to protect citizens and the environment. Furthermore, if successful, this model would also reduce costs and manpower in the long term for HAB screening and monitoring.

## Cost of Errors

Incorrectly predicting the presence and severity of Harmful Algal Blooms (HABs) can have negative effects on the ecosystem and the community. If the blooms are not accurately predicted, there is a risk of underestimating the potential harm they can cause. This can lead to a lack of preparation and response to the blooms, allowing them to persist and worsen.

Underestimating the impact of HABs can result in the continued exposure of humans, pets, and marine life to toxic substances, causing harm to their health and potentially leading to fatalities. Additionally, industries such as aquaculture and agriculture can be severely impacted by the persistence of HABs, leading to decreased productivity and potential financial losses.

Accurate predictions of HABs are crucial in order to ensure the safety of the community and the preservation of the ecosystem. Incorrect predictions can result in a lack of proper response and action, allowing HABs to cause harm and destabilize the environment.

# Understanding the Data

Data is available from all agencies, the EPA has manually sampled over 23,500 sites across the United States from 2013 through 2021 and provided severity levels based on Cynanobacteria count for over 17,000 sites. **Pulling the Imagery from Planetary Computer**

```python
catalog = Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1", 
    modifier=pc.sign_inplace
)
```

```python
# search the planetary computer sentinel-l2a and landsat level-2 collections
search = catalog.search(
    collections=["sentinel-2-l2a", "landsat-c2-l2"],
    bbox=bbox,
    datetime=date,
    # This lets me limit photos to those with cloud cover less than specified amount
    # for this pull, I will not limit cloud cover
    query={'eo:cloud_cover': {'lt':100}}
)
```

```python
# Creating a list of all of the search results
search_items = [item for item in search.get_all_items()]
len(search_items)
```

```python
# Checking the properties of a specific search result
search_items[5].properties
```

**Pulling All Satellite Images from Sample Search**

Now that I have my search results, I can create a dataframe with information that will be useful to this analysis.

```python
pic_details = []
for pic in search_items:
    pic_details.append(
    {
    'item': pic,
    'satelite_name':pic.collection_id,
    'img_date':pic.datetime.date(),
    'cloud_cover(%)': pic.properties['eo:cloud_cover'],
    'img_bbox': pic.bbox,
    'min_long': pic.bbox[0],
    "max_long": pic.bbox[2],
    "min_lat": pic.bbox[1],
    "max_lat": pic.bbox[3]
    }
    )
```

```python
pd.DataFrame(pic_details).head()
```

**Checking That the Sample Point is Within the Satellite Image**

For our search the bound box pulled any samples for the specified date range within a 31 mile buffer zone. I need to ensure that the sample location is actually captured within the satellite image.

```python
big_buffer_df = pd.DataFrame(pic_details)

big_buffer_df['has_sample_point'] = (
        (big_buffer_df.min_lat < sample_row.latitude)
        & (big_buffer_df.max_lat > sample_row.latitude)
        & (big_buffer_df.min_long < sample_row.longitude)
        & (big_buffer_df.max_long > sample_row.longitude)
    )

big_buffer_df = big_buffer_df[big_buffer_df['has_sample_point'] == True]
big_buffer_df.head(3)
```

```python
len(big_buffer_df)
```

Only 7 images out of an original 45 actually include the coordinates of the sample location.

## Elevation Data

Pulling the elevation data for the same 500 random data points. This is also a lengthy process and takes about 5 minutes.

```python
# selecting only relevant features
elev_random = random_batch[['uid','latitude', 'longitude', 'tiny_crop_bbox']]
```

```python
# creating new coordinate feature and dropping now irrelevant features
elev_random['coords'] = list(zip(elev_random['longitude'], elev_random['latitude']))
elev_random = elev_random.drop(['longitude', 'latitude'], axis=1)
```

```python
# Creating a new column to hold the collected elevation data
elev_random['elevation'] = np.NaN
```

```python
elev_random.head()
```

```python
# pulling in the data and adding it to the dataframe
catalog = Client.open(
"https://planetarycomputer.microsoft.com/api/stac/v1", modifier=pc.sign_inplace
)

for index in range(len(elev_random)):

    # selects the next row regardless of index
    row = elev_random.iloc[index]
    
    # Pulling in the elevation data from Copernicus
    search = catalog.search(
    collections=["cop-dem-glo-30"],
    intersects={"type": "Point", "coordinates": row.coords},
    )
    items = list(search.get_items())

    
    # setting smaller boundary for elevation data
    (minx, miny, maxx, maxy) = row.tiny_crop_bbox

    signed_asset = pc.sign(items[0].assets["data"])
    data = (
        rioxarray.open_rasterio(signed_asset.href).rio.clip_box(
            minx=minx,
            miny=miny,
            maxx=maxx,
            maxy=maxy,
            crs="EPSG:4326",
        )
        )
    # adding the average elevation to the dataframe at specified index
    elev_random['elevation'].loc[row.name] = data.to_numpy().mean()
```

```python
elevation_df = elev_random.copy()
elevation_df.head()
```

## Cleaning DataFrame and Finalizing Features

Now I'll be demonstrating the process used to clean the DataFrame and engineer some new features by using the random sample that was pulled in from above. 

Steps taken include:
- Dropping irrelevant columns
- Creating a column to keep track of the difference from the satellite image date to the sample date
    - For rows that do not have an image, filling in the image date with the sample date and creating an indicator column
- Converting dates to ordinal 
- OneHotEncoding seasons

Later, I will impute any additional features that are missing from the DataFrame.

```python
# reseting the index
full_df_indexed = random_batch_features.reset_index().drop('index', axis=1)
full_df_indexed.tail()
```

Dropping Irrelevant columns.

```python
model_df = full_df_indexed[['date', 'uid', 'region', 'latitude', 'longitude', 'season', 
                            'img_date', 'red_mean', 'red_median', 'red_max', 
                            'red_min','red_sum', 'red_product', 'green_mean', 
                            'green_median', 'green_max', 'green_min', 'green_sum',
                            'green_product', 'blue_mean', 'blue_median','blue_max',
                            'blue_min', 'blue_sum', 'blue_product', 'severity']]
```

Checking for nulls

```python
model_df.isna().sum()
```

I need to fill in the img_date column and will use the value from the date column. To make sure that the dataframe knows that there was missing info, I will add an indicator for the rows with missing img_date.

```python
model_df[model_df['img_date'].isnull()].head()
```

```python
missing_index_list = list(model_df[model_df['img_date'].isnull()].index)
```

```python
model_df['missing_img_date'] = 0
```

```python
for index in missing_index_list:
    model_df['missing_img_date'].loc[index] =1
```

Now that I have my missing indicator column, I will fill the missing img_dates with the sample date.

```python
model_df['img_date'] = model_df['img_date'].fillna(model_df['date'])
```

Now I need to correct all of the dtypes  and make a new column for the difference between sample date and image date.

```python
model_df.info()
```

Converting the dtypes to the correct type.

```python
model_df['date'] = model_df['date'].apply(lambda x: datetime.date(x))
model_df['latitude'] = model_df['latitude'].apply(lambda x: x.astype(float))
model_df['longitude'] = model_df['longitude'].apply(lambda x: x.astype(float))
model_df['severity'] = model_df['severity'].apply(lambda x: x.astype(int))
model_df['date'] = model_df['date'].apply(lambda x: x.toordinal())
model_df['img_date'] = model_df['img_date'].apply(lambda x: x.toordinal())
model_df['days_from_sat_to_sample'] = model_df['date'] - model_df['img_date']
```

```python
model_df.head()
```

One Hot Encoding the seasons category. There is no data leakage from converting string labels to integers.

```python
# One hot encoding seasons
ohe = OneHotEncoder(sparse=False)
seasons = ohe.fit_transform(model_df[['season']])
cols = ohe.get_feature_names_out()
# converting new ohe to dataframe
seasons = pd.DataFrame(seasons, columns=cols, index=model_df.index)
model_df = pd.concat([model_df.drop('season', axis=1), seasons], axis=1)
```
